```r
Question 3:
Simulate a white noise series with 1000 random draws and plot (1) a time series plot and (2) a histogram.
The usual reqirements on graphics (described) in Question 1) applied.
```

```r
w=rnorm(1000,0,1) # Make 1000 independent random draw from a standard normal distribution

#1)
plot.ts(w, main="Simulated White Noise", col="navy",
          ylab="Simulated values", xlab="Simulated Time Period") #looks cyclic around the same mean and variance

#2)
hist(w,main="Simulated White Noise", col="blue",
     xlab="Simulated Values")
```

```r
length(w)
z <- w
# We are going to construct a simulated AR(1) model manually instead of
# using built-in function from R
for (t in 2:length(w)){
  z[t] <- 0.9*z[t-1] + w[t] # use the same random normal sequence generated above
}
```

```r
hist(z, breaks="FD",
     main="AR(ar=c(0.9))",
     xlab="Values of a Simluated Zero-Mean AR(1) Series",
     col="blue", labels=TRUE)

plot.ts(z, main="Simulated AR(ar=c(0.9)) Series", col="navy",
        ylab="Values of the Simluated Series",
        xlab="Simulated Time Period")

#It has a smoother appearence, and the cyclyes can be identfied.
```

```r
z <- w
# We are going to construct a simulated AR(1) model manually instead of
# using built-in function from R
for (t in 2:length(w)){
  z[t] <- 0.2*z[t-1] + w[t] # use the same random normal sequence generated above
}

hist(z, breaks="FD",
     main="AR(ar=c(0.2))",
     xlab="Values of a Simluated Zero-Mean AR(1) Series",
     col="blue", labels=TRUE)

plot.ts(z, main="Simulated AR(ar=c(0.2)) Series", col="navy",
        ylab="Values of the Simluated Series",
        xlab="Simulated Time Period")
```

```r
Simulate (with 1000 random draws) the following 3 models:

1.A deterministic linear (time) trend of the form: yt=10+0.5t
2. Random walk without drift
3. Random walk with drift = 0.5
```

```r
# Random walk with zero drift
x=cumsum(w) 

# Random walk with drift = 0.5
wd = 0.5 + w; 
xd = cumsum(wd) 

# Check out the numbers to see if they make sense
#head(cbind(w,x,wd,xd),20)

# Set graphic parameters
#GRAPH_BLUE<-rgb(43/255, 71/255,153/255)
#par(bg="grey95") # Background color for the plot

par(mfrow=c(1,1))
plot.ts(xd, main="Random Walk with Drift, Random Walk without Drift, Deterministic Trend",
        col="blue", ylab="Values", xlab="Simulated Time Period", bg=38)
lines(10+.5*(1:length(x)), lty="dashed", col="navy")
lines(x, col="purple")
  # Add vertical lines
  abline(v=c(100,200,300,400),col=3,lty=3)
  # Add Legend
  leg.txt <- c("RW with Drift" , "Deterministic Linear Trend", "RW without Drift")
  legend("topleft", legend=leg.txt, lty=c(1,2,1), col=c("blue","navy","purple"),
         bty='n', cex=1, merge = TRUE, bg=336)

#Random walk with drift seem more along the lie of DLT
```
